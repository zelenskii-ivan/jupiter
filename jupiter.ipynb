```python
from sympy import *
from sympy.plotting import plot as plt

init_printing()

x = Symbol('x')
fx = -12*x**4*sin(cos(x)) - 18*x*3+5*x*2 + 10*x - 30
```

```python
solveset(diff(fx) > 0)
```

```python
solveset(diff(fx) < 0)
```

```python
p = plt(fx)
print(p)
```

```python
solveset(fx>0)
```

```python
solveset(fx<0)
```
